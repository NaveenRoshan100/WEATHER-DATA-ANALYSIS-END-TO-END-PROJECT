```python
import pandas as p
import requests as r
import json
from datetime import datetime
import os
```

CURRENT DATA -WEATHERAPI.COM

```python
api=os.environ['weather_api']
today=datetime.today().strftime('%Y-%M-%D')
url=f'http://api.weatherapi.com/v1/forecast.json?key={api}&q=chennai&dt={today}'
data=r.get(url)
data=data.json()
```

```python
forcast = p.json_normalize(data['forecast']['forecastday'][0]['hour'], sep='.')
forcast.to_csv('forcast.csv', index=False)

astro = p.json_normalize(data['forecast']['forecastday'][0]['astro'], sep='.')
astro.to_csv('astro.csv', index=False)
```

HISTORY DATA -OPEN METEO

```python

lat, lon = 13.0833, 80.2833

url = "https://archive-api.open-meteo.com/v1/era5"

params = {
    "latitude": lat,
    "longitude": lon,
    "start_date": "2015-01-01",  
    "end_date": "2025-09-13",
    "hourly": [
        "temperature_2m",
        "relative_humidity_2m",
        "dewpoint_2m",
        "apparent_temperature",
        "surface_pressure",
        "cloudcover",
        "windspeed_10m",
        "winddirection_10m",
        "precipitation",
        "weathercode"
        
    ],
    "timezone": "Asia/Kolkata"
}

resp = r.get(url, params=params)
data = resp.json()
df = p.DataFrame(data["hourly"])
```
